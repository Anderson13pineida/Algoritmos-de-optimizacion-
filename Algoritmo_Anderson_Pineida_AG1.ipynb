2.Desarrollo de algoritmo voraz para resolver problemas(devolución de cambio)


```python
def cambio_monedas(N, SM): #SM: sistema monetario son las denominaciones de las monedas disponibles para dar el cambio
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(28,[25,10,5,1])
```

3.Desarrollo de algoritmo con la técnica de vuelta atrás(backtracking)(N-Reinas)


```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

4.Desarrollo de algoritmo con Programación dinámica(paseo por el rio)

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

Dado un conjunto de puntos se trata de encontrar los dos puntos más cercanos <b>

# Paso 1: Fuerza Bruta (1D)
- Este enfoque compara todos los pares de puntos y encuentra el par más cercano.

```python
# Fuerza bruta en 1D
def closest_points_brute_force(points):
    n = len(points)
    if n < 2:
        return None, float('inf')  # No hay pares de puntos

    min_distance = float('inf')
    closest_pair = None

    for i in range(n):
        for j in range(i + 1, n):
            dist = abs(points[i] - points[j])
            if dist < min_distance:
                min_distance = dist
                closest_pair = (points[i], points[j])

    return closest_pair, min_distance

# Ejemplo
points_1d = [3403, 4537, 9089, 9746, 7259]
result_1d = closest_points_brute_force(points_1d)
print("Fuerza Bruta 1D:", result_1d)
```

# Paso 2: Divide y Vencerás (1D)
- Mejoramos la solución dividiendo los puntos y combinando los resultados.

```python
# Divide y vencerás en 1D
def closest_points_divide_and_conquer(points):
    def merge_and_find_closest(left, right, delta):
        mid = (left + right) // 2
        mid_point = sorted_points[mid]

        # Encontrar puntos en la banda dentro de delta
        strip = [p for p in sorted_points[left:right + 1] if abs(p - mid_point) < delta]
        strip.sort()  # Ordenar por posición

        min_distance = delta
        closest_pair = None

        # Comparar pares en la banda
        for i in range(len(strip)):
            for j in range(i + 1, len(strip)):
                if abs(strip[i] - strip[j]) >= min_distance:
                    break
                dist = abs(strip[i] - strip[j])
                if dist < min_distance:
                    min_distance = dist
                    closest_pair = (strip[i], strip[j])
        return closest_pair, min_distance

    def closest_rec(left, right):
        if right - left + 1 <= 3:
            # Fuerza bruta en pequeños subproblemas
            return closest_points_brute_force(sorted_points[left:right + 1])

        mid = (left + right) // 2
        left_pair, left_min = closest_rec(left, mid)
        right_pair, right_min = closest_rec(mid + 1, right)

        delta = min(left_min, right_min)
        cross_pair, cross_min = merge_and_find_closest(left, right, delta)

        if cross_min < delta:
            return cross_pair, cross_min
        return (left_pair if left_min <= right_min else right_pair), delta

    sorted_points = sorted(points)
    return closest_rec(0, len(sorted_points) - 1)

# Ejemplo
result_1d_divide_conquer = closest_points_divide_and_conquer(points_1d)
print("Divide y Vencerás 1D:", result_1d_divide_conquer)
```
